This cell will create an array x ranging from 0 to 1 in 100 steps

```python
import matplotlib.pyplot as plot
import numpy as np
x = np.linspace(0,1,100)
```

This cell defines two functions that return sin(x) and cos(x) respectively

```python
def ReturnSin(input):
    return np.sin(input)
def ReturnCos(input):
    return np.cos(input)
```

This cell plots the graphs of sin(x) and cos(x) in a multipanel plot

```python
fig, axs = plot.subplots(ncols=2)
axs[0].plot(x)
axs[0].plot(ReturnSin(x))
axs[0].set(xlabel="sin(x) versus x")
axs[1].plot(x)
axs[1].plot(ReturnCos(x))
axs[1].set(xlabel = "cos(x) versus x")

```
