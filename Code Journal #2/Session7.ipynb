This cell defines an array x that ranges from 0,1 with 100 steps

```python
import matplotlib.pyplot as matt
import numpy as np

x = np.linspace(0,1,100)
```

This cell defines a function that returns exp(X)

```python
def exponent(ex):
    return np.exp(ex)
```

This cell sets y=exp(x) and plots x versus y.

```python
y = exponent(x)
matt.plot(x)
matt.plot(y)
matt.xlabel("Time [milliseconds]")
matt.ylabel("Awesomeness")
matt.show
```
