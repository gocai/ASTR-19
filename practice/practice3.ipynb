```python
import matplotlib.pyplot as plot
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(5 * x)
```

```python
plot.plot(x)
plot.plot(y)
plot.ylim([-1.1,1.1])
plot.xlabel("X axis")
plot.ylabel("Y axis")
plot.gca().set_facecolor("g")
plot.savefig('demo.png', facecolor='g')
plot.show
```
